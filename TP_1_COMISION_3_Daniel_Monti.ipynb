**EJERCICIO 3 CICLO FOR**


Solicitar al usuario que ingrese 10 números. Calcule la suma de todos los números ingresados y  cuente cuántos de ellos son par. Mostrar la suma de todos los numeros y mostrar cuanto de ellos son par.


```python
#Escribe aqui abajo la resolucion del EJERCICIO 3:

print('Ingrese 10 numeros')
suma = 0
contador = 0

for i in range(9):
    numero=int(input("Numero: "))
    suma += numero
    if numero % 2 == 0:
        contador += 1

print("Suma de todos los numeros ingresados:", suma)
print("Cantidad de numeros par", contador)
```

**EJERCICIO 7 VECTORES CON CICLO FOR**

Solicitar al usuario que ingrese una X cantidad de elementos a un vector(lista).

Ingresar cada elemento del vector mediante input() y calcular el promedio de dicho vector.

AYUDA: EL PROMEDIO ES LA SUMA DE TODOS LOS NÚMEROS DIVIDO LA CANTIDAD TOTAL.

Por ultimo mostrar el vector.

```python
#Escribe aqui abajo la resolucion del EJERCICIO 7:
suma = 0
lista_numeros = []

print('Ingrese 5 numeros')

for i in range(5):
    nro=int(input())
    lista_numeros.append(nro)

for numero in lista_numeros:
    suma += numero

print(suma)
print("Promedio de la lista", suma/5)
```

**EJERCICIO 8 LISTA CON CICLO FOR**

Solicitar al usuario que ingrese 10 personas con los siguiente datos en formato lista.

['NOMBRE','APELLIDO', 'DNI','GASTOS MENSUALES DE HOGAR']

Ingresar al final de la lista estos datos, de 2 nuevas personas:

['ESTEBAN','LOPEZ', '45675400','190800.80']

['NOELIA','ALEJA', '46906875','85900.50']


Eliminar la primera persona de la lista.

Mostrar el vector completo.

```python
#Escribe aqui abajo la resolucion del EJERCICIO 8:

print("Ingresar 10 personas la siguiente informacion: ")
listaPersonas = []
for i in range(9):
    persona = []
    nombre = input("Ingrese Nombre:")
    apellido = input("Ingrese Apellido:")
    dni = input("Ingrese DNI:")
    gastos=input("Ingrese Gastos Mensuales:")

    persona.append(nombre)
    persona.append(apellido)
    persona.append(dni)
    persona.append(gastos)

    listaPersonas.append(persona)

listaPersonas.append(['ESTEBAN','LOPEZ', '45675400','190800.80'])
listaPersonas.append(['NOELIA','ALEJA', '46906875','85900.50'])

#Borrar elemento 0 de ListaPersonas
listaPersonas.pop(0)

#Muestra resultado final
print("******Lista completa********")
for personas in listaPersonas:
    print(personas)
```
