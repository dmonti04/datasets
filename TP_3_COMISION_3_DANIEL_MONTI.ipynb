***APLICACIÓN DE PROBABILIDAD***

Se Realiza el lanzamiento de un dado EQUILIBRADO JUSTO, de manera reiterada con las mismas condiciones, y se anotan sus resultados= CARA O CRUZ.

Se le pide al usuario que defina la cantidad N de lanzamientos y se calcula su frecuencia relativa= 'PROBABILIDAD' de obtener cara o cruz. Se obtiene los siguientes resultados:

```python
import numpy as np
n=int(input('ingrese la cantidad de simulación o lanzamientos'))

#========= FORMA 1 DE SIMULAR LANZAMIENTO DE UN DADO============
dado=[]
cara=0
cruz=0
for i in range(n):
  dado.append(np.random.rand(1))
  dado1=np.array(dado)
  if dado1[i]<0.5:
    #print('salio cara')
    #print(dado[i])
    cara +=1
  else:
    #print('salio cruz')
    #print(dado[i])
    cruz +=1
```

```python
#  mostrar la probailidad frecuencista de el lanzamiento de un dado
prom_cara=cara/n
prom_cruz=cruz/n
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CARA es:', prom_cara)
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CRUZ es:', prom_cruz)
```

```python
#========== FORMA COMPACTA DE SIMULAR USANDO NUMPY===========
m=int(input('ingrese la cantidad de simulación o lanzamientos'))
lanzamientos = np.random.randint(0, 2, m)

# Calcular las probabilidades

probabilidad_cara = np.sum(lanzamientos ==0 ) / len(lanzamientos)
probabilidad_sello = np.sum(lanzamientos== 1) / len(lanzamientos)

print(f"Probabilidad de cara: {probabilidad_cara:.2f}")
print(f"Probabilidad de sello: {probabilidad_sello:.2f}")
```

```python
import matplotlib.pyplot as plt
x=np.array(range(2))
y=np.array([probabilidad_cara, probabilidad_sello])

plt.bar(x,y)
plt.xlabel('Opción')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma del lanzamiento de una moneda equilibrada')
plt.show()
```

##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
#Importar librerias

import numpy as np
import matplotlib.pyplot as plt

# Crear el vector lanzamiento de un dado equilibrado de 6 caras
lanzamientos = np.random.randint(1, 7, 100) #Genera 100 numeros aleatorios de 1 a 6 enteros

# Cargar en un vector booleano que asuma cada elemento el valor True si el valor generado
# por numpy es igual a cada valor del dado, es decir si salieron 23 veces 1 se cargara este
# vector con 23 True, el resto de los elementos sera False

prob1_bool = lanzamientos == 1
prob2_bool = lanzamientos == 2  #lanzamiento[2, 7, 2, 3..] -->prob2_bool=[True, False, True, False]
prob3_bool = lanzamientos == 3
prob4_bool = lanzamientos == 4
prob5_bool = lanzamientos == 5
prob6_bool = lanzamientos == 6

# Contar las veces que salio cada valor

prob1_contador = np.sum(prob1_bool)
prob2_contador = np.sum(prob2_bool)
prob3_contador = np.sum(prob3_bool)
prob4_contador = np.sum(prob4_bool)
prob5_contador = np.sum(prob5_bool)
prob6_contador = np.sum(prob6_bool)

# Calculo de la probabilidad = numero de ocurrencias verdaderas sobre el total de
# Lanzamientos

prob1= prob1_contador/len(lanzamientos)
prob2= prob2_contador/len(lanzamientos)
prob3= prob3_contador/len(lanzamientos)
prob4= prob4_contador/len(lanzamientos)
prob5= prob5_contador/len(lanzamientos)
prob6= prob6_contador/len(lanzamientos)

# mostrar resultados de las probabilidades de 2, 3 y 5

print(f"Probabilidad de obtener 2: {prob2: .2f}")
print(f"Probabilidad de obtener 3: {prob3: .2f}")
print(f"Probabilidad de obtener 5: {prob5: .2f}")

# Mostrar graficos de las probabilidades

#Rango de numeros a mostrar y probabilidades
numeros = [1, 2, 3, 4, 5, 6]
probabilidades = [prob1, prob2, prob3, prob4, prob5, prob6]

plt.bar(numeros, probabilidades)
plt.xlabel('Numeros')
plt.ylabel('Probabilidad')
plt.title('Histograma de probabilidades')
plt.show()





```

## EJERCICIO 3

 Realizar la simulación del juego {piedra,papel , tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probailidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidad. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
#Importar librerias

import numpy as np
import matplotlib.pyplot as plt

# Crear el vector jugadas del juego piedra, papel o tijera. con tres posibilidades
#  1 cuando elige 'piedra'
#  2 cuando elige 'papel'
#  3 cuando elige 'tijera'

jugadas = np.random.randint(1, 4, 10) #Genera 10 numeros aleatorios de 1 a 3 enteros

# Cargar en un vector booleano que asuma cada elemento el valor True si el valor generado
# por numpy es igual 1 = 'papel', 2 = 'papel' y 3 = 'tijera'

prob1_bool = jugadas == 1
prob2_bool = jugadas == 2
prob3_bool = jugadas == 3


# Contar las veces que salio cada valor

prob1_contador = np.sum(prob1_bool)
prob2_contador = np.sum(prob2_bool)
prob3_contador = np.sum(prob3_bool)


# Calculo de la probabilidad = numero de ocurrencias verdaderas sobre el total
# de jugadas

prob1= prob1_contador/len(jugadas)
prob2= prob2_contador/len(jugadas)
prob3= prob3_contador/len(jugadas)

# mostrar resultados de las probabilidades de elegir piedra cuando sale el 1,
# papel cuando sale el 2 y tijera cuando sale el 3

print(f"Probabilidad de elegir 'piedra': {prob1: .2f}")
print(f"Probabilidad de elegir 'papel' : {prob2: .2f}")
print(f"Probabilidad de elegir 'tigera': {prob3: .2f}")

# Mostrar graficos de las probabilidades

#Rango de numeros a mostrar y probabilidades
eligio = ['piedra', 'papel', 'tijera']
probabilidades = [prob1, prob2, prob3]

plt.bar(eligio, probabilidades)
plt.xlabel('Elección')
plt.ylabel('Probabilidad')
plt.title('Histograma de probabilidades')
plt.show()
```

**Conclusión:** Al aumentar el numero de jugadas la probabilidad que elija 'piedra', 'papel' o 'tijera' tienden a ser iguales

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas diarias de 4 productos durante una semana

ventas_diarias = np.array(
    #Lun,Mar,Mie,Jue,Vie,Sab,Dom
    [[20, 15, 25, 30, 18, 22, 24],  #Producto A
    [12, 20, 14, 8, 15, 18, 16],    #Producto B
    [35, 28, 32, 30, 26, 24, 30],   #Producto C
    [40, 38, 45, 42, 39, 41, 37]    #Producto D
    ]
)
# 1 Calcular la media en la ventas por cada producto y la mediana.
#   Plotear estos nuevos datos en un gráfico de barra
ventas_orden_por_fila= np.sort(ventas_diarias, axis=1)

media_ventas_producto= np.mean(ventas_orden_por_fila, axis = 1)
mediana_ventas_producto= np.median(ventas_orden_por_fila, axis = 1)

print(f"Media de las ventas por cada producto: {media_ventas_producto}")
print(f"Mediana de las ventas por cada producto: {mediana_ventas_producto}\n")

# Mostrar graficos de barras

x = ['producto A', 'producto B', 'producto C','producto D']
y1 = media_ventas_producto
y2 =  mediana_ventas_producto

plt.bar(x, y1, label='Media por producto', alpha=0.5)
plt.bar(x, y2, label='Mediana por producto', alpha=0.5)
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.legend(loc = 'upper left')
plt.title('Media y Mediana de las ventas por producto')
plt.show()

# 2 Calcular la media en la ventas por DÍA y la mediana.
# Plotear estos nuevos datos en un gráfico de TRAZO DE LINEAS.
ventas_orden_por_col= np.sort(ventas_diarias, axis=0)

media_ventas_dia= np.mean(ventas_orden_por_col, axis = 0)
mediana_ventas_dia= np.median(ventas_orden_por_col, axis = 0)

print("\n")
print(f"Media de las ventas por dia: {media_ventas_dia}")
print(f"Mediana de las ventas por dia: {mediana_ventas_dia}\n")

# Mostrar graficos

x = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
y1 = media_ventas_dia
y2 =  mediana_ventas_dia

plt.plot(x, y1, label= 'Media', alpha=0.5)
plt.plot(x, y2, label= 'Mediana', alpha=0.5)
plt.xlabel('Dias')
plt.ylabel('Ventas')
plt.legend(loc = 'upper right')
plt.title('Media y Mediana de las ventas por dias')
plt.show()

# 4 - Calcular el desvio estandar de los datos por cada producto y concluir
# para cada producto como es su rango de variabilidad en las ventas de los mismos

# Plotear los datos de ventas de la semana de los productos con mayor y menor desvio estandar
# y evidenciar la variabilidad indicada en el parametro 'DESVIO ESTANDAR'

desviacion_estandar = np.std(ventas_diarias, axis=1)

# Obtener valor=indice del producto com Mayor y Menor desvio estandar
max_desviacion= np.argmax(desviacion_estandar)
min_desviacion= np.argmin(desviacion_estandar)
print("\n\n")
# Mostrar graficos de barras

x = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
y1 = ventas_diarias[max_desviacion]
y2 = ventas_diarias[min_desviacion]
plt.plot(x, y1, label='Mayor variabilidad')
plt.plot(x, y2, label='Menor variabilidad')
plt.xlabel('Dias de la semana')
plt.ylabel('Ventas')
plt.legend(loc = 'upper right')
plt.title('Mayor y menor Variabilidad de ventas por productos de la semana ')
plt.show()





```

2.   Los encuestados son de la provincia de Cordoba. Además el 80 % de la población de esa provincia tiene un grupo familiar de 5 personas. Con ingresos mensuales promedio de $258.000
2.   Cerca del 80 % de los encuestados Gastan el 90% de sus ingresos, el resto gasta el 80% de sus ingresos.Ayuda, para simular la cantidad de miembros utilizar la distribución normal.
3.  Del gasto Total por hogar se sabe que se abona un 15 % en telefonía y un 7.8% en servicio de internet.

Con la información proporcionada, realizar la simulación de las encuestas de 100 y 1000 personas.

```python
import numpy as np
import matplotlib.pyplot as plt

eleccion= int(input("elija una 1: para simular 100 personas o 2: para simular 1000 personas"))

if eleccion==1:
    nro_encuestados=100
elif eleccion==2:
    nro_encuestados=1000

m = nro_encuestados

encuestadosCordoba = np.array([])

# 1. Genercion edades y DNI

edades = np.random.choice(['Jovenes', 'Mayores', 'Adultos'], nro_encuestados,  p=[0.7, 0.1, 0.2])

# Generacion DNI

jovenes_bool = edades == 'Jovenes'
adultos_bool = edades == 'Adultos'
mayores_bool = edades == 'Mayores'

cantidad_jovenes= np.sum(jovenes_bool)
cantidad_adultos= np.sum(adultos_bool)
cantidad_mayores= np.sum(mayores_bool)

dni_jovenes= np.arange(40000000, 40000000 + cantidad_jovenes, 1)
dni_adultos= np.arange(20000000, 20000000 + cantidad_adultos, 1)
dni_mayores= np.arange(4000000, 4000000 + cantidad_mayores, 1)

dni= np.concatenate((dni_jovenes, dni_adultos, dni_mayores), axis=0)

# Simulación de grupos familiares y sus ingresos

grupos_familiares = np.random.normal(5, 1, nro_encuestados ) #5= media, 1= desviacion estandar
grupos_familiares_int= grupos_familiares.astype(int)

# Simulación de grupos ingresos
# 258000= media, 100000= desviacion estandar supuesta

ingresos_mensuales =  np.random.normal(258000, 100000, nro_encuestados)

'''
 Simulacion gastos mensuales
 np.random.rand(nro_encuestados) < 0.8, esta es la condicion que usaremos para np.where
 np.random.rand(nro_encuestados genera numeros aleatorios del 0 al 1, los numeros < 0.8
 los numeros < 0.8 (true en el where) representan el del 0 al 80 % de los encunestados y estos gastan el 90 % --> ingresos_mensuales * 0.9
 los numeros > 0.8 (false en where) representan el del 0 al 20 % de los encunestados y estos gastan el 80 % --> ingresos_mensuales * 0.8
'''

gastos_mensuales = np.where(np.random.rand(nro_encuestados) < 0.8, ingresos_mensuales * 0.9, ingresos_mensuales * 0.8)

# Del gasto Total el 15 % se gasta en telefonía

gasto_Mensual_telefonia= gastos_mensuales * 0.15

# Del gasto Total el 7.8% se gasta en servicio de internet.

gasto_mensual_internet=gastos_mensuales * 0.078


# Pago de servicios la suma de gastos de telefonia e internet

pago_servicios = gasto_Mensual_telefonia + gasto_mensual_internet

# Construccion de la matriz poblacionCordoba

encuestadosCordoba= np.concatenate((dni, ingresos_mensuales, grupos_familiares_int, gasto_Mensual_telefonia, gasto_mensual_internet), axis=0)
matriz_encuestadosCordoba= np.transpose( encuestadosCordoba.reshape(5, m))
print(f"\nMatriz de {m} personas encuestadas\n{matriz_encuestadosCordoba}")

# Calcular el valor medio y desvio estandar de los gastos mensuales, ingresos mensuales y pagos de servicios.

media_gastos_mensuales = np.mean(gastos_mensuales)
desvio_gastos_mensuales = np.std(gastos_mensuales)

media_ingreso_mensual = np.mean(ingresos_mensuales)
desvio_ingreso_mensual = np.std(ingresos_mensuales)

media_telefonia = np.mean(gasto_Mensual_telefonia)
desvio_telefonia = np.std(gasto_Mensual_telefonia)

media_internet = np.mean(gasto_mensual_internet)
desvio_internet = np.std(gasto_mensual_internet)

# 2. Plotear los datos de los ingresos , y gastos en gráficos de barra por separado
# y evidenciar la media y desvio estandar.


plt.bar(['Telefonía', 'Internet'], [media_telefonia, media_internet], yerr=[desvio_telefonia, desvio_internet], capsize=10)
plt.ylabel('Gasto en pesos')
plt.xlabel('Servicios')
plt.title("Media y Desvio estandar de pago de servicios")
plt.show()

print("\n\n")

plt.bar(['Ingresos'], [media_ingreso_mensual], yerr=[desvio_ingreso_mensual], capsize=10)
plt.ylabel('Pesos')
plt.title("Media y desvio en pesos de ingresos mensual")
plt.show()

# 3. Calcular la proporción de jóvenes menores a 21, esos jovenes
#   cantidad_jovenes = edades == 'Jovenes' uso esta expresion en 1.
proporcion_jovenes = np.sum(cantidad_jovenes) / nro_encuestados

print(f"\nProporción de jóvenes menores a 21 para una muestra (N={nro_encuestados}): {proporcion_jovenes:.2f}\n")

# 4. Histograma de grupos familiares
plt.hist(grupos_familiares, bins=20, alpha=0.7, label= f'Grupos Familiares {nro_encuestados}')

# Mostrar cantidad de miembros, frecuencia e histograma
plt.xlabel("Cantidad de Miembros")
plt.ylabel("Frecuencia")
plt.title("Histograma de Grupos Familiares")
plt.legend()
plt.show()
```
