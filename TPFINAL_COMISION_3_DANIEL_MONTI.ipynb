El dataset elegido, "All Countries and their Economies", contiene datos relacionados al comportamiento económico de todos los paises. Este dataset trata de mostrar un paneo general de cómo es la economia en cada país.


1. **Country:** se refiere al nombre del país en estudio.

2. **Poverty headcount ratio at \$2.15 a day(2017 PPP)(% of population):**
se refiere a la proporción de personas, medida en porcentaje de la población total, que vive por debajo del umbral de pobreza de $2.15 al día ajustado al poder adquisitivo (PPP). Puede ser un indicador de la pobreza monetaria o económica de la región.

3. **Life expectancy at birth, total (years):**
se refiere al promedio de años que se espera que una persona viva desde el momento de su nacimiento en una población o región específica. Puede ser una medida importante para evaluar la calidad de vida de la región.

4. **Population, total:**
representa la cantidad total de individuos que viven en una región o área geográfica específica. Puede ser importante para entender la demografía y la composición de la población.

5. **Population growth (annual %):**
se refiere a la tasa de crecimiento anual de la población de una región, expresada en porcentaje. Puede interpretarse de la siguiente manera:

   *  Si el valor es positivo, indica un crecimiento poblacional.
   *  Si el valor es cero, significa que no hubo cambios respecto al año anterior.
   *  Si el valor es negativo, indica una disminución de la población.

6. **Net migration:**
representa la diferencia entre el número de personas que ingresan al país (inmigración) y el número de personas que lo abandonan (emigración):

   *  Si el valor es positivo, indica que están ingresando más personas de las que salen.
   *  Si el valor es cero, indica que la cantidad de personas que ingresan y abandonan el país es la misma.
   *  Si el valor es negativo, indica que están abandonando el país más personas de las que ingresan.

7. **Human Capital Index (HCI) (scale 0-1):**
se refiere al índice de capital humano, evalúa el nivel de habilidades y conocimientos de la población en una región o país. Se mide en una escala que va desde 0 hasta 1, donde 0 representa un bajo capital humano y 1 representa un capital humano máximo o muy alto.

8. **GDP:** se refiere al Producto Interno Bruto (PIB) del país.  El PIB es una medida del valor total de todos los bienes y servicios producidos.

9. **GDP per capita:** se refiere al Producto Interno Bruto (PIB) del país dividido por su población total. Representa la cantidad de riqueza económica que, en promedio, está disponible para cada habitante de esa área.

10. **GDP growth (annual %):**  se refiere a la tasa de crecimiento anual del Producto Interno Bruto (PIB) del país, expresada como un porcentaje.

   *  Si el valor es positivo, indica un crecimiento económico.
   *  Si el valor es muy cercano a cero, indica un estancamiento.
   *  Si el valor es negativo, indica un decrecimiento de la economía.

11. **Unemployment, total (% of total labor force) (modeled ILO estimate):**
se refiere a la tasa de desempleo total del país, expresada como un porcentaje del total de la fuerza laboral.

12. **Inflation, consumer prices (annual %):**
se refiere a la tasa de inflación anual del país, expresada como un porcentaje.

13. **Personal remittances, received (% of GDP):**
se refiere al porcentaje de las remesas personales recibidas por un país en relación con su Producto Interno Bruto (PIB). Las remesas personales son transferencias de dinero realizadas por individuos que trabajan en el extranjero y las envían a sus familiares o personas en su país de origen.

14. **CO2 emissions (metric tons per capita):**
se refiere a las emisiones de dióxido de carbono (CO2) del país por habitante, medidas en toneladas métricas por persona.

15. **Forest area (% of land area):**
se refiere al porcentaje de la superficie terrestre del país que está cubierto por bosques. Esta medida cuantifica la proporción de tierra que está ocupada por bosques en relación con el área total de tierra.

  *  Un valor más alto en esta variable indica que una mayor proporción del territorio está cubierta por bosques.
  *  Un valor más bajo indica que una menor proporción del territorio está cubierta por bosques.

16. **Access to electricity (% of population):**
 se refiere al porcentaje de la población que tiene acceso a la electricidad. Esta variable se utiliza para evaluar el progreso en la electrificación de una población y la inclusión de la población en la economía moderna.

17. **Annual freshwater withdrawals, total (% of internal resources):**
se refiere al porcentaje del total de extracciones de agua dulce anuales en el país con respecto a los recursos internos disponibles de agua dulce. Esta medida cuantifica la cantidad de agua dulce que se extrae anualmente en relación con la cantidad de agua dulce disponible en la región.

18. **Electricity production from renewable sources, excluding hydroelectric (% of total):**
se refiere al porcentaje de la producción total de electricidad que proviene de fuentes de energías renovables, pero excluyendo la energía hidroeléctrica.

19. **People using safely managed sanitation services (% of population):**
se refiere al porcentaje de la población que tiene acceso a servicios de saneamiento gestionados de manera segura.

20. **Intentional homicides (per 100,000 people):**
se refiere a la tasa de homicidios intencionales por cada 100.000 habitantes. Esta medida cuantifica la cantidad de homicidios intencionales, es decir, asesinatos deliberados y no accidentales, en relación con la población total.

21. **Central government debt, total (% of GDP):**
se refiere al porcentaje de la deuda total del gobierno central en relación con su Producto Interno Bruto (PIB).

   *  Un valor más alto en esta variable indica que la deuda del gobierno central es proporcionalmente mayor en relación con el tamaño de la economía del país, lo que puede indicar una mayor carga de deuda y desafíos financieros.
   *  Un valor más bajo indica que la deuda del gobierno central es proporcionalmente menor en relación con el PIB, lo que generalmente se considera más favorable desde el punto de vista de la salud financiera del país.
   
  Esta variable es importante para evaluar la solidez financiera del gobierno central y su capacidad para manejar la deuda pública.

22. **Statistical performance indicators (SPI): Overall score (scale 0-100):**
se refiere a un indicador de rendimiento estadístico que evalúa la calidad y precisión de los datos estadísticos recopilados y publicados por una entidad o agencia gubernamental. Este indicador se mide en una escala que va desde 0 hasta 100, donde 0 representa un bajo nivel de calidad en la producción y publicación de datos estadísticos, y 100 representa la máxima calidad y precisión en la generación de datos estadísticos.

23. **Individuals using the Internet (% of population):**
 se refiere al porcentaje de la población que utiliza Internet.

24. **Proportion of seats held by women in national parliaments (%):**
se refiere al porcentaje de puestos en el parlamento nacional de un país que están ocupados por mujeres.

25. **Foreign direct investment, net inflows (% of GDP):**
se refiere al porcentaje del Producto Interno Bruto (PIB) de un país que representa la entrada neta de inversión extranjera directa (IED) en ese país. Se habla de inversión extranjera directa cuando una entidad extranjera, como una empresa o una inversión individual, adquiere una participación significativa en una empresa o activo en otro país.

Detectar Valores Ausente y Valores Atípicos. Decidir si eliminarlos y el por qué de la elección.

Este código proporciona una visión general de la presencia de valores ausentes y valores atípicos en un conjunto de datos numérico. La creación de DataFrames para visualizar estos resultados ayuda a entender mejor la calidad y la distribución de los datos, lo que es crucial para el análisis posterior.

```python
import numpy as np
import pandas as pd

# Cargar los datos desde la URL
url = 'https://gitlab.com/Exequiel.Mollo/dataset_paises_economias/-/raw/main/All_Countries_and_Economies.csv'
datos = pd.read_csv(url)

# En esta sección del código, se crea un DataFrame llamado resultados_ausentes
    # que muestra la cantidad de valores ausentes y el porcentaje de valores ausentes en cada columna del conjunto de datos.
# Los valores se obtienen a través de la variable valores_ausentes,
    # que almacena la cantidad de valores nulos en cada columna, y porcentaje_ausentes, que almacena el porcentaje correspondiente.


# Creación del DataFrame para Valores Ausentes:
valores_ausentes = datos.isna().sum()

# Calcular el porcentaje de valores ausentes por columna
porcentaje_ausentes = (valores_ausentes / len(datos)) * 100

# Crear un DataFrame para visualizar los resultados de los valores ausentes
resultados_ausentes = pd.DataFrame({'Valores Ausentes': valores_ausentes, 'Porcentaje Ausentes': porcentaje_ausentes})
print("Valores Ausentes por Columna:")
print(resultados_ausentes)


# Cálculo de Cuartiles y Rango Intercuartil:
#Aquí se calculan los cuartiles (percentiles 25 y 75) y el rango intercuartil (IQR) de las columnas numéricas en el conjunto de datos.
#Esto proporciona una medida de la dispersión y variabilidad de los datos numéricos.

# Calcular los cuartiles y el rango intercuartil (IQR) de cada columna numérica
cuartil1 = datos.select_dtypes(include=np.number).quantile(0.25)
cuartil3 = datos.select_dtypes(include=np.number).quantile(0.75)
rango_intercuartil = cuartil3 - cuartil1

# Definición de Límites para Valores Atípicos:
# Se definen límites utilizando los cuartiles y el rango intercuartil.
# La idea es identificar valores que se alejan significativamente de la tendencia central de los datos.
# Los límites se establecen en función de una regla empírica, utilizando 1.5 veces el IQR.
limite_inferior = cuartil1 - 1.5 * rango_intercuartil
limite_superior = cuartil3 + 1.5 * rango_intercuartil

# Detección de Valores Atípicos:
# En esta parte del código, se verifica si cada valor en las columnas numéricas está por debajo del límite inferior o por encima del límite superior definidos previamente.
# Esto resulta en una matriz booleana (valores_atipicos) que muestra dónde se encuentran los valores atípicos.
valores_atipicos = (datos.select_dtypes(include=np.number) < limite_inferior) | (datos.select_dtypes(include=np.number) > limite_superior)

# Cuenta la cantidad de valores atípicos por columna
conteo_valores_atipicos = valores_atipicos.sum()

# Creación del DataFrame para Valores Atípicos:
# Se crea otro DataFrame llamado resultados_atipicos que muestra la cantidad de valores atípicos en cada columna.
# Esto se logra sumando los valores True en la matriz booleana valores_atipicos para cada columna.
resultados_atipicos = pd.DataFrame({'Valores Atípicos': conteo_valores_atipicos})
print("\nValores Atípicos por Columna:")
print(resultados_atipicos)


```

Si observamos la tabla de "Valores Ausentes por Columna", podemos notar que diferentes columnas tienen diferentes porcentajes de valores ausentes. Algunas columnas tienen muy pocos valores ausentes (cercanos al 0%), mientras que otras tienen porcentajes más altos.

Este código elimina las filas que contienen los valores atípicos en las columnas específicas definidas en columnas_atipicas del DataFrame df_cleaned. Luego, se muestra información sobre los tipos de variables y la dimensión del DataFrame limpio resultante.

```python
# Definición de la función remove_outliers:
# Esta función toma un DataFrame como argumento y se encarga de eliminar las filas que contienen valores atípicos en cualquiera de las columnas.
# Utiliza el rango intercuartil (IQR) para determinar qué valores se consideran atípicos y cuáles no.

# q1, q3 = df[column].quantile([0.25, 0.75]): Calcula el primer cuartil (Q1) y el tercer cuartil (Q3) de los valores en la columna actual.
# iqr = q3 - q1: Calcula el rango intercuartil (IQR) restando Q1 de Q3.
# df = df[~((df[column] < q1 - 1.5 * iqr) | (df[column] > q3 + 1.5 * iqr))]:
# Utiliza la operación booleana para seleccionar las filas que están dentro del rango intercuartil (no son valores atípicos) y asigna el DataFrame resultante nuevamente a df.
# El operador ~ se utiliza para invertir la condición, seleccionando las filas que están fuera del rango intercuartil (valores atípicos).

# Llamada a la función remove_outliers:
# Después de haber definido la función, se llama a esta función utilizando el DataFrame df_cleaned como argumento.
# Esto eliminará los valores atípicos del DataFrame limpio.

#Mostrar información del DataFrame limpio:
#Finalmente, se imprime información sobre el DataFrame resultante después de eliminar los valores atípicos.
#Esto incluye la información sobre los tipos de variables en cada columna y la dimensión (número de filas y columnas) del DataFrame resultante.

def remove_outliers(df):
    """Elimina las filas del dataframe que contienen valores atípicos.

    Args:
        df: El dataframe a limpiar.

    Returns:
        El dataframe sin valores atípicos.
    """

    for column in df.columns:
        # Calcula los límites inferior y superior del rango intercuartil.
        q1, q3 = df[column].quantile([0.25, 0.75])
        iqr = q3 - q1

        # Elimina las filas que se encuentran fuera de los límites intercuartiles.
        df = df[~((df[column] < q1 - 1.5 * iqr) | (df[column] > q3 + 1.5 * iqr))]

    return df

# Llama a la función para eliminar valores atípicos del DataFrame limpio
df_cleaned = remove_outliers(df_cleaned)

# Mostrar información sobre los tipos de variables y dimensiones del DataFrame limpio
print('\n')
print('\033[34m\n■ Tipo de variable que tengo en cada columna en el DataFrame limpio\033[33m\n')
print(df_cleaned.info())
print('\n')
print('\033[34m\n■ Dimensión del DataFrame limpio:\033[33m', f' ►► {df_cleaned.shape}')
print('\n')
```

**Ejercicio 7** Sobre el Dataset elegido explicar si se puede reducir las dimensiones y qué representan esas nuevas variables.

Luego de haber aplicado el algoritmo para encontrar las Componentes principales (CP) vemos que si es posible una reducción de la dimensionalidad, mi conjunto de datos antes de aplicar PCA, tenían 13 variables (caracteristicas) y se pudo reducir a 7 Componentes principales, estas componentes principales son nueva variables que se crean como combinaciones lineales de las variables originales.

Esta técnica de la estadística me permite, simplificar y representar datos complejos en un espacio de menor dimensionalidad, manteniendo la mayor cantidad posible de la variabilidad original. En el proceso de encontrar las Componentes Principales el criterio de tener una varianza acumulada de más de 0.8 me asegura la  captura de la mayor varianza posible. Dependiendo del problema es probable que pueda con pocas componentes principales lograr esta captura, pero habrá casos que elegiré tomar un valor de 0.7 de varianza acumulada lo que me llevara a perder la variabilidad original, pero lograre la reducción de la dimensionalidad lo cual facilita la interpretación de resultados y sacar conclusiones.


```python
# PCA: Analisis de componentes principales
# ESTANDARIZACION de datos

column_name= df_cleaned.columns.values  #.columns.values: devuelve los nombres de las columnas  ¡¡¡COPIADO DE PUNTO MULTIVARIADO!!

df_cuntries_economies= df_cleaned[column_name]

scaler= StandardScaler()
df_cuntries_economies_scaled= scaler.fit_transform(df_cuntries_economies)

# Aplicar PCA: Aplicar el calculo de los componentes principales
pca= PCA(n_components= None)

#Usar los datos ajustados para hallar relaciones y hacer el cálculo de componentes principales
pca.fit(df_cuntries_economies_scaled)

#______________________________________________________________________________
# Calculamos la varianza explicada y la varianza explicada acumulativa
'''
La varianza explicada (también llamada variación explicada )
se usa para medir la discrepancia entre un modelo y los datos reales
explained_variance_ratio_ : nos devuelve el porcentaje de varianza explicada por cada componente principal:
'''
varianza_explicada = pca.explained_variance_ratio_
varianza_explicada_acumulada = np.cumsum(varianza_explicada)
'''
La varianza_explicada_acumulada me sirve para ver cuando llego a 1, pero una varianza
acumada explicada que llega a 0.8 o 0.85 puedo considerar como optima cuando llegue
a estos valores podre saber cuantas componentes principales puedo tomar.
'''

#Graficar la Varianza explicada acumulativa

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_explicada_acumulada)+1), varianza_explicada_acumulada, marker='o')

plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza explicada acumulativa(sumada)')

plt.title('Resultados del PCA')
plt.show()


#En base al resultado del gráfico "Resultados del PCA", elegimos retener 7 componentes principales, dado es el menor número de componentes con el que
#podemos obtener una varianza explicada acumulativa (sumada) igual o mayor a 0.8.

#Número de componentes principales a CAPTURAR o RETENER
numeros_de_componentes_a_capturar= 7

#Aplicar entonces ahora PCA para 7 componentes principales:
pca= PCA(n_components= numeros_de_componentes_a_capturar)
pca.fit(df_cuntries_economies_scaled)

#_______________________________________________________________________________
#Análisis e interpretación de los componentes principales
'''
.componentes_: devuelve una matriz donde el numero de filas es igual a la cantidad
 de componentes que decidimos capturar y el valor que encontramos en cada columna
 representa la importancia de esa caracteristica
'''
componentes= pca.components_

# Visualizar la importancia de las características en el primer componente principal

plt.figure(figsize=(10, 6))
plt.bar(range(len(componentes[0])), componentes[0])  # Dibuja una barra para cada característica
plt.xticks(range(len(componentes[0])),column_name)   # Etiquetas en el eje X
plt.xticks(rotation=90)                              # Gira 90° las etiquetas del eje X
plt.xlabel('Características')
plt.ylabel('Importancia')
plt.title('Importancia de Características en el Primer Componente Principal')
plt.show()


# Obtener la lista de características con sus contribuciones a cada componente (cuánto contribuye cada característica al modelo)
feature_contributions=pd.DataFrame(componentes,columns= column_name)  # Crea tabla para mostrar las contribuciones
print("\nContribuciones de las Características en Cada Componente Principal:\n")
print(feature_contributions)  # Muestra las contribuciones

```
